```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
lst_marks = [42, 55, 67, 78, 89, 90, 45, 60, 72, 88.5, 95, 100, 45, 55, 65, 75, 85, 95, 100]
mimimum, q1, median, q3, maximum = np.quantile(lst_marks, [0, 0.25, 0.5, 0.75, 1.0])
```

```python
mimimum, q1, median, q3, maximum 
```

```python
np.mean(lst_marks), np.std(lst_marks)
```

```python
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = [x for x in lst_marks if x < lower_bound or x > upper_bound]
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
print(f"Outliers: {outliers}")
```

```python
sns.boxplot(lst_marks)
plt.title("Box Plot of Marks")
plt.xlabel("Marks")
```
