```python
def square(n):
    for i in range(n):
        yield i ** 2
```

```python
result = square(5)
```

```python
next(result)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen = my_generator()
```

```python
next(gen)
```

```python
## Reading large files with generators

def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
for line in read_large_file('large_file.txt'):
    print(line)  # Process each line as needed
```
