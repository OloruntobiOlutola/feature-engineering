### Feature Engineering Process

```python
df_flight['Date'] = df_flight['Date_of_Journey'].str.split('/').str[0].astype(int)
df_flight['Month'] = df_flight['Date_of_Journey'].str.split('/').str[1].astype(int)
df_flight['Year'] = df_flight['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
df_flight.head(5)
```

```python
df_flight.info()
```

```python
# Drop the original 'Date_of_Journey' column
df_flight.drop(columns=['Date_of_Journey'], inplace=True)
```

```python
df_flight.head()
```

```python
# Extract hour
df_flight['Arrival_hour'] = df_flight['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0].astype(int)

# Extract minute
df_flight['Arrival_minute'] = df_flight['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1].astype(int)
```

```python
df_flight.head()
```

```python
df_flight.drop(columns=['Arrival_Time'], inplace=True)
```

```python
df_flight
```

```python
df_flight['Depature_hour'] = df_flight['Dep_Time'].str.split(':').str[0].astype(int)
df_flight['Depature_minute'] = df_flight['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
df_flight.head(5)
```

```python
df_flight.drop(columns=['Dep_Time'], inplace=True)
```

```python
df_flight['Total_Stops'].isna().sum()
```

```python
df_flight['Total_Stops'] = df_flight['Total_Stops'].map({
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4,
    np.nan: 1
})
```
